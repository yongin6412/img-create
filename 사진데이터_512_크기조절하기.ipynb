```python
import os
import glob
from PIL import Image

files = glob.glob('/content/drive/MyDrive/Colab_Notebooks/shoes/stylegan2-ada-pytorch/datasetsAll/shoseall/*.jpg') #jpg확장자로 끝나는 모든파일을 files에 넣음

for f in files:
    title, ext = os.path.splitext(f)  #파일경로 확장자확인
    if ext in ['.jpg']:
        img = Image.open(f)
        img_resize = img.resize((512, 512)) #사진크기를 512X512으로 만든다
        img_resize.save(title+ext)
```
